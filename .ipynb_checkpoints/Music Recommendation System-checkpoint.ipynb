```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
df['User-Rating']
```

```python
l=[]
for i in df['User-Rating']:
    l.append(i[:3])
l
```

```python
df['User-Rating']=l
df
```

```python
df['Album/Movie']=df['Album/Movie'].str.replace(' ',' ')
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ',' ')
df
```

```python
df['Singer/Artists']=df['Singer/Artists'].str.replace(' ',' ')
df
```

```python
df['tags']=df['Singer/Artists']+' '+df['Genre']+' '+df['Album/Movie']+' '+df['User-Rating']
df['tags'][1:5]
```

```python
new_df = df[['Song-Name','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000)
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
cv.vocabulary_
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity[0][0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
new_df.rename(columns={'Song-Name':'title'},inplace=True)
```

```python
def recommend(music):
    music_index=new_df[new_df['title']==music].index[0]
    distances=similarity[music_index]
    music_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in music_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Tere Bin')
```

```python
recommend('Proper Patola')
```
